### Now let's visualize our data

```python
#plotting graph category vs sub-category
plt.figure(figsize=(18,8))
plt.bar('Sub-Category','Category', data=df)
plt.xlabel('Sub-Category',size=18)
plt.ylabel('Category',size=18)
plt.show()
```

```python
print(df['State'].value_counts())
plt.figure(figsize=(15,8))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.show()
```

```python
print(df['Sub-Category'].value_counts())
plt.figure(figsize=(15,8))
sns.countplot(x=df['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

```python
#category/sub-category vs sales/profit
df.groupby(['Category','Sub-Category'])['Profit','Sales'].sum().plot.bar(figsize=(10,8))
plt.title('Total profit and sales per sub-category')
plt.legend(loc = 'upper center')
plt.show()
```

##### Here we can see that PHONES AND CHAIRS have maximum selling price, whereas maximum profit is earned in COPIERS.

```python
df.groupby('State')['Profit','Sales'].sum().plot.bar(figsize=(15,10))
plt.show()
```

##### Here we can see that instead of extremely high selling prices of the products in California, profit is also maximum in the California and NewYork. According to this data we can assume that rich people lives in California and NewYork.

```python
df.groupby('Category')['Profit','Sales'].sum().plot.bar(figsize=(15,10))
plt.show()
```

#### According to this graph we can see that PROFIT is maximum in TECHNOLOGY and minimum in FURNITURE.

```python
plt.figure(figsize=(17,10))
sns.countplot(x='Region',hue='Ship Mode',data=df)
plt.show()
```
